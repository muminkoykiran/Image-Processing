```python
import numpy as np
import matplotlib.pyplot as plt
data_path = "/home/mumin/Image-Processing/6.hafta/"
im_1=plt.imread(data_path+"denizli.jpg")
im_1.shape
```

```python
im_2=np.zeros((im_1.shape[0],im_1.shape[1]),dtype=np.uint8)
im_2.shape
im_2=im_1[:,:,0]
im_3=np.zeros((im_1.shape[0],im_1.shape[1]),dtype=np.uint8)
#im_3=im_1[:,:,0]
plt.imshow(im_2,cmap='gray')
plt.show()
```

```python
m,n=im_2.shape
```

```python
for i in range(1,m-1):
    for j in range(1,n-1):
        s = im_2[i-1,j-1]/9 + im_2[i-1,j]/9 + im_2[i-1,j+1]/9 + im_2[i,j-1]/9 + im_2[i,j]/9 + im_2[i,j+1]/9 + im_2[i+1,j-1]/9 + im_2[i+1,j]/9 + im_2[i+1,j+1]/9
        s=int(s)
        im_3[i,j]=s
```

```python
plt.subplot(1,2,1)
plt.imshow(im_2,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(im_3,cmap='gray')
```

```python
plt.imsave(data_path + "test4.jpg", im_2,cmap='gray')
plt.imsave(data_path + "test4-1.jpg", im_3,cmap='gray')
```
